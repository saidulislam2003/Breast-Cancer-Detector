# Breast Cancer Prediction

## Data Cleaning and EDA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Breast_cancer_dataset.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop(['id'], axis = 1, inplace = True)
```

```python
data.info()
```

```python
data = data.dropna(axis=1, how='all')
```

```python
data.info()
```

```python
data['diagnosis'] = data['diagnosis'].map({'B': 0, 'M': 1})
data['diagnosis'].value_counts()
```

```python
data.describe()
```

## Visualization

```python
correlation = data.corr()
correlation
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(correlation, cmap = 'coolwarm', annot = False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x = 'diagnosis', data=data)
plt.title("Malignant vs Benign Count")
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.show()
```

```python
selected_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'diagnosis']
sns.pairplot(data[selected_features], hue='diagnosis')
plt.show()
```

```python
import matplotlib.pyplot as plt

for col in selected_features[:-1]:
    sns.boxplot(x='diagnosis', y=col, data = data)
    plt.title(f'{col} vs Diagnosis')
    plt.show()
```

## Train Test Split and Model Development

```python
X = data.drop('diagnosis', axis = 1)
y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 42)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 1000, random_state = 42)
```

```python
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

## Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(kernel = 'rbf', C = 1, gamma = 'scale')
model_svm.fit(X_train, y_train)
```

```python
y_pred_svm = model_svm.predict(X_test)
y_pred_svm
```

```python
accuracy_score(y_test, y_pred_svm)
```
